**О наборе данных**

Этот набор данных предоставляет детальный обзор тренировочных привычек посетителей тренажерного зала, их физических характеристик и показателей фитнеса. Он содержит 973 записи данных из тренажерного зала, включая ключевые показатели эффективности, такие как частота пульса, сожженные калории и продолжительность тренировки. Каждая запись также включает демографические данные и уровень опыта, что позволяет проводить комплексный анализ фитнес-паттернов, прогресса атлетов и тенденций в состоянии здоровья.

**Ключевые характеристики:**

*   **Возраст (Age):** Возраст посетителя тренажерного зала.
*   **Пол (Gender):** Пол посетителя (Мужчина или Женщина).
*   **Вес (кг) (Weight (kg)):** Вес посетителя в килограммах.
*   **Рост (м) (Height (m)):** Рост посетителя в метрах.
*   **Макс. пульс (Max_BPM):** Максимальная частота пульса (ударов в минуту) во время тренировок.
*   **Ср. пульс (Avg_BPM):** Средняя частота пульса во время тренировок.
*   **Пульс в покое (Resting_BPM):** Частота пульса в состоянии покоя до тренировки.
*   **Продолжительность тренировки (часы) (Session_Duration (hours)):** Длительность каждой тренировки в часах.
*   **Сожженные калории (Calories_Burned):** Общее количество калорий, сожженных за тренировку.
*   **Тип тренировки (Workout_Type):** Тип выполненной тренировки (например, Кардио, Силовая, Йога, ВИИТ).
*   **Процент жира (Fat_Percentage):** Процентное содержание жира в организме посетителя.
*   **Потребление воды (литры) (Water_Intake (liters)):** Ежедневное потребление воды в дни тренировок.
*   **Частота тренировок (дни/неделю) (Workout_Frequency (days/week)):** Количество тренировок в неделю.
*   **Уровень опыта (Experience_Level):** Уровень опыта, от начинающего (1) до эксперта (3).
*   **ИМТ (BMI):** Индекс массы тела, рассчитанный на основе роста и веса.

Этот набор данных идеально подходит для специалистов по анализу данных, исследователей в области здоровья и фитнес-энтузиастов, заинтересованных в изучении привычек тренировок, моделировании фитнес-прогресса или анализе взаимосвязи между демографическими и физиологическими данными. Благодаря широкому спектру переменных, он позволяет получить представление о том, как различные факторы влияют на интенсивность тренировок, выносливость и общее состояние здоровья.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

def read_and_process_data(filename):
    """
    Чтение и обработка CSV файла с данными о тренировках
    """
    # Чтение данных из CSV файла
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        data = list(reader)
    
    # Извлечение числовых данных
    numeric_data = {}
    text_data = {}
    
    for field in reader.fieldnames:
        values = []
        for row in data:
            try:
                # Пробуем преобразовать в число
                value = float(row[field])
                values.append(value)
            except ValueError:
                # Если не число, сохраняем как текст
                if field not in text_data:
                    text_data[field] = []
                text_data[field].append(row[field])
        
        if values:  # Если есть числовые значения
            numeric_data[field] = np.array(values)
    
    return numeric_data, text_data, reader.fieldnames

def calculate_statistics(numeric_data):
    """
    Расчет статистик для числовых полей
    """
    stats = {}
    
    for field, values in numeric_data.items():
        stats[field] = {
            'mean': np.mean(values),
            'std': np.std(values),
            'min': np.min(values),
            'max': np.max(values),
            'median': np.median(values)
        }
    
    return stats

def plot_histograms(numeric_data, stats):
    """
    Построение гистограмм для числовых полей
    """
    n_fields = len(numeric_data)
    cols = 4
    rows = (n_fields + cols - 1) // cols
    
    plt.figure(figsize=(20, 5 * rows))
    
    for i, (field, values) in enumerate(numeric_data.items()):
        plt.subplot(rows, cols, i + 1)
        
        # Гистограмма
        n, bins, patches = plt.hist(values, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
        
        # Добавление линий для среднего и СКО
        mean = stats[field]['mean']
        std = stats[field]['std']
        
        plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Среднее: {mean:.2f}')
        plt.axvline(mean + std, color='orange', linestyle='--', linewidth=1, label=f'+1 СКО: {mean + std:.2f}')
        plt.axvline(mean - std, color='orange', linestyle='--', linewidth=1, label=f'-1 СКО: {mean - std:.2f}')
        
        plt.title(f'Гистограмма: {field}')
        plt.xlabel(field)
        plt.ylabel('Частота')
        plt.legend()
        plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def calculate_correlations(numeric_data):
    """
    Расчет корреляций между полями
    """
    # Создаем матрицу данных для расчета корреляций
    fields = list(numeric_data.keys())
    n_fields = len(fields)
    n_samples = len(numeric_data[fields[0]])
    
    # Создаем матрицу данных
    data_matrix = np.zeros((n_samples, n_fields))
    
    for i, field in enumerate(fields):
        data_matrix[:, i] = numeric_data[field]
    
    # Расчет корреляционной матрицы
    correlation_matrix = np.corrcoef(data_matrix.T)
    
    return correlation_matrix, fields

def plot_correlation_heatmap(correlation_matrix, fields):
    """
    Построение тепловой карты корреляций
    """
    fig, ax = plt.subplots(figsize=(12, 10))
    
    im = ax.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
    
    # Добавляем подписи
    ax.set_xticks(np.arange(len(fields)))
    ax.set_yticks(np.arange(len(fields)))
    ax.set_xticklabels(fields, rotation=45, ha='right')
    ax.set_yticklabels(fields)
    
    # Добавляем значения в ячейки
    for i in range(len(fields)):
        for j in range(len(fields)):
            text = ax.text(j, i, f'{correlation_matrix[i, j]:.2f}',
                          ha="center", va="center", color="black", fontsize=8)
    
    plt.colorbar(im)
    plt.title('Матрица корреляций между полями')
    plt.tight_layout()
    plt.show()

def print_statistics(stats):
    """
    Вывод статистик в читаемом формате
    """
    print("=" * 80)
    print("СТАТИСТИЧЕСКИЕ ХАРАКТЕРИСТИКИ ПОЛЕЙ")
    print("=" * 80)
    
    for field, stat in stats.items():
        print(f"\n{field}:")
        print(f"  Среднее значение: {stat['mean']:.2f}")
        print(f"  СКО: {stat['std']:.2f}")
        print(f"  Минимум: {stat['min']:.2f}")
        print(f"  Максимум: {stat['max']:.2f}")
        print(f"  Медиана: {stat['median']:.2f}")

def find_strong_correlations(correlation_matrix, fields, threshold=0.7):
    """
    Поиск сильных корреляций между полями
    """
    print(f"\nСИЛЬНЫЕ КОРРЕЛЯЦИИ (|r| > {threshold}):")
    print("=" * 50)
    
    strong_correlations = []
    
    for i in range(len(fields)):
        for j in range(i + 1, len(fields)):
            corr = correlation_matrix[i, j]
            if abs(corr) > threshold:
                strong_correlations.append((fields[i], fields[j], corr))
    
    # Сортируем по абсолютному значению корреляции
    strong_correlations.sort(key=lambda x: abs(x[2]), reverse=True)
    
    for field1, field2, corr in strong_correlations:
        direction = "положительная" if corr > 0 else "отрицательная"
        strength = "сильная" if abs(corr) > 0.8 else "умеренная"
        print(f"{field1} ↔ {field2}: {corr:.3f} ({strength} {direction} корреляция)")
    
    return strong_correlations

def find_weak_correlations(correlation_matrix, fields, threshold=0.1):
    """
    Поиск слабых корреляций между полями
    """
    print(f"\nСЛАБЫЕ КОРРЕЛЯЦИИ (|r| < {threshold}):")
    print("=" * 50)
    
    weak_correlations = []
    
    for i in range(len(fields)):
        for j in range(i + 1, len(fields)):
            corr = correlation_matrix[i, j]
            if abs(corr) < threshold:
                weak_correlations.append((fields[i], fields[j], corr))
    
    # Сортируем по абсолютному значению корреляции
    weak_correlations.sort(key=lambda x: abs(x[2]))
    
    # Выводим только топ-10 самых слабых корреляций
    for field1, field2, corr in weak_correlations[:10]:
        direction = "положительная" if corr > 0 else "отрицательная"
        print(f"{field1} ↔ {field2}: {corr:.3f} (очень слабая {direction} корреляция)")
    
    return weak_correlations

def plot_scatter_matrix(numeric_data, strong_correlations, weak_correlations):
    """
    Построение матрицы графиков рассеивания для интересных пар переменных
    """
    # Выбираем наиболее интересные пары для визуализации
    interesting_pairs = []
    
    # Добавляем несколько самых сильных корреляций
    for field1, field2, corr in strong_correlations[:4]:
        interesting_pairs.append((field1, field2, corr, 'strong'))
    
    # Добавляем несколько самых слабых корреляций
    for field1, field2, corr in weak_correlations[:4]:
        interesting_pairs.append((field1, field2, corr, 'weak'))
    
    # Добавляем несколько случайных пар для сравнения
    fields = list(numeric_data.keys())
    np.random.seed(42)
    random_indices = np.random.choice(len(fields), 4, replace=False)
    for i in range(0, len(random_indices), 2):
        if i+1 < len(random_indices):
            field1 = fields[random_indices[i]]
            field2 = fields[random_indices[i+1]]
            corr = np.corrcoef(numeric_data[field1], numeric_data[field2])[0, 1]
            interesting_pairs.append((field1, field2, corr, 'random'))
    
    # Строим матрицу графиков
    n_pairs = len(interesting_pairs)
    cols = 4
    rows = (n_pairs + cols - 1) // cols
    
    plt.figure(figsize=(20, 5 * rows))
    
    for idx, (field1, field2, corr, corr_type) in enumerate(interesting_pairs):
        plt.subplot(rows, cols, idx + 1)
        
        # Цвет в зависимости от типа корреляции
        if corr_type == 'strong':
            color = 'red' if corr > 0 else 'blue'
            alpha = 0.6
        elif corr_type == 'weak':
            color = 'gray'
            alpha = 0.3
        else:
            color = 'green'
            alpha = 0.4
        
        # График рассеивания
        plt.scatter(numeric_data[field1], numeric_data[field2], 
                   alpha=alpha, color=color, s=20)
        
        # Линия тренда
        if abs(corr) > 0.2:  # Добавляем линию тренда только если есть какая-то корреляция
            z = np.polyfit(numeric_data[field1], numeric_data[field2], 1)
            p = np.poly1d(z)
            plt.plot(np.sort(numeric_data[field1]), p(np.sort(numeric_data[field1])), 
                    color='black', linewidth=2, linestyle='--')
        
        plt.title(f'{field1} vs {field2}\nкорреляция: {corr:.3f}')
        plt.xlabel(field1)
        plt.ylabel(field2)
        plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def analyze_correlation_results(correlation_matrix, fields, strong_correlations, weak_correlations):
    """
    Анализ результатов корреляционного анализа
    """
    print("\n" + "=" * 80)
    print("АНАЛИЗ КОРРЕЛЯЦИОННОГО АНАЛИЗА")
    print("=" * 80)
    
    print("\n1. ВЫСОКОКОРРЕЛИРОВАННЫЕ АТРИБУТЫ:")
    print("-" * 40)
    
    # Группируем сильные корреляции по типу
    positive_strong = [(f1, f2, c) for f1, f2, c in strong_correlations if c > 0]
    negative_strong = [(f1, f2, c) for f1, f2, c in strong_correlations if c < 0]
    
    if positive_strong:
        print("\nПоложительные высокие корреляции:")
        for field1, field2, corr in positive_strong[:5]:  # Показываем топ-5
            print(f"  • {field1} и {field2}: {corr:.3f}")
            print(f"    - Интерпретация: при увеличении {field1} увеличивается {field2}")
    
    if negative_strong:
        print("\nОтрицательные высокие корреляции:")
        for field1, field2, corr in negative_strong[:5]:
            print(f"  • {field1} и {field2}: {corr:.3f}")
            print(f"    - Интерпретация: при увеличении {field1} уменьшается {field2}")
    
    print("\n2. АТРИБУТЫ БЕЗ КОРРЕЛЯЦИИ:")
    print("-" * 40)
    
    # Находим пары с наименьшей корреляцией
    weak_pairs = []
    for i in range(len(fields)):
        for j in range(i + 1, len(fields)):
            corr = correlation_matrix[i, j]
            if abs(corr) < 0.1:
                weak_pairs.append((fields[i], fields[j], corr))
    
    weak_pairs.sort(key=lambda x: abs(x[2]))
    
    print(f"Найдено {len(weak_pairs)} пар с очень слабой корреляцией (|r| < 0.1)")
    if weak_pairs:
        print("Примеры таких пар:")
        for field1, field2, corr in weak_pairs[:5]:
            print(f"  • {field1} и {field2}: {corr:.3f}")
            print(f"    - Это означает, что изменения в {field1} практически не влияют на {field2}")
    
    print("\n3. ВЫВОДЫ И ИНТЕРПРЕТАЦИЯ:")
    print("-" * 40)
    print("""
    • Сильные положительные корреляции обычно указывают на взаимосвязанные показатели 
      (например, рост и вес, продолжительность тренировки и сожженные калории)
    
    • Сильные отрицательные корреляции могут указывать на компенсаторные механизмы
      или обратные зависимости
    
    • Отсутствие корреляции между параметрами может означать:
      - Независимость процессов
      - Разные физиологические механизмы
      - Нелинейную зависимость, не улавливаемую коэффициентом корреляции Пирсона
      - Наличие скрытых переменных, влияющих на связь
    
    • Для фитнес-данных ожидаемы следующие корреляции:
      - Положительные: вес ↔ ИМТ, длительность тренировки ↔ сожженные калории
      - Отрицательные: ИМТ ↔ уровень опыта, возраст ↔ максимальный пульс
    """)
    
    # Анализ конкретных фитнес-корреляций
    print("\n4. СПЕЦИФИЧЕСКИЕ ФИТНЕС-КОРРЕЛЯЦИИ:")
    print("-" * 40)
    
    # Проверяем конкретные пары, интересные для фитнес-анализа
    fitness_pairs = [
        ('Calories_Burned', 'Session_Duration (hours)', 'Сожженные калории и продолжительность тренировки'),
        ('BMI', 'Weight (kg)', 'ИМТ и вес'),
        ('Max_BPM', 'Age', 'Максимальный пульс и возраст'),
        ('Experience_Level', 'Fat_Percentage', 'Уровень опыта и процент жира'),
        ('Workout_Frequency (days/week)', 'Calories_Burned', 'Частота тренировок и сожженные калории')
    ]
    
    for field1_name, field2_name, description in fitness_pairs:
        if field1_name in fields and field2_name in fields:
            idx1 = fields.index(field1_name)
            idx2 = fields.index(field2_name)
            corr = correlation_matrix[idx1, idx2]
            
            print(f"\n{description}:")
            print(f"  Корреляция: {corr:.3f}")
            
            if abs(corr) > 0.7:
                print(f"  Очень сильная корреляция")
            elif abs(corr) > 0.5:
                print(f"  Сильная корреляция")
            elif abs(corr) > 0.3:
                print(f"  Умеренная корреляция")
            elif abs(corr) > 0.1:
                print(f"  Слабая корреляция")
            else:
                print(f"  Практически отсутствует корреляция")
            
            if corr > 0:
                print(f"  Положительная зависимость: при увеличении {field1_name} увеличивается {field2_name}")
            elif corr < 0:
                print(f"  Отрицательная зависимость: при увеличении {field1_name} уменьшается {field2_name}")

def main():
    filename = 'gym_members_exercise_tracking.csv'
    
    try:
        # Чтение и обработка данных
        print("Чтение данных...")
        numeric_data, text_data, fieldnames = read_and_process_data(filename)
        
        # Расчет статистик
        print("Расчет статистик...")
        stats = calculate_statistics(numeric_data)
        
        # Вывод статистик
        print_statistics(stats)
        
        # Построение гистограмм
        print("\nПостроение гистограмм...")
        plot_histograms(numeric_data, stats)
        
        # Расчет и визуализация корреляций
        print("\n" + "=" * 80)
        print("КОРРЕЛЯЦИОННЫЙ АНАЛИЗ")
        print("=" * 80)
        
        print("\nРасчет корреляций...")
        correlation_matrix, fields = calculate_correlations(numeric_data)
        
        print("Построение тепловой карты корреляций...")
        plot_correlation_heatmap(correlation_matrix, fields)
        
        # Поиск сильных корреляций
        strong_correlations = find_strong_correlations(correlation_matrix, fields)
        
        # Поиск слабых корреляций
        weak_correlations = find_weak_correlations(correlation_matrix, fields)
        
        # Построение матрицы графиков рассеивания
        print("\nПостроение матрицы графиков рассеивания...")
        plot_scatter_matrix(numeric_data, strong_correlations, weak_correlations)
        
        # Анализ результатов
        analyze_correlation_results(correlation_matrix, fields, strong_correlations, weak_correlations)
        
        # Дополнительная информация о данных
        print(f"\n" + "=" * 80)
        print("ОБЩАЯ ИНФОРМАЦИЯ О ДАННЫХ")
        print("=" * 80)
        print(f"Количество записей: {len(numeric_data[list(numeric_data.keys())[0]])}")
        print(f"Количество числовых полей: {len(numeric_data)}")
        print(f"Количество текстовых полей: {len(text_data)}")
        
        if text_data:
            print("\nТекстовые поля:")
            for field, values in text_data.items():
                unique_values = set(values)
                print(f"  {field}: {len(unique_values)} уникальных значений")
                if len(unique_values) <= 10:
                    print(f"    Значения: {', '.join(unique_values)}")
        
    except FileNotFoundError:
        print(f"Ошибка: Файл '{filename}' не найден!")
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```
